```python
import matplotlib.pyplot as plt

# Data for the methods and their metrics
methods = ["Semantic Search", "BM25", "Query Decomposition", "Hybrid RAG", "HyDE"]
rouge_l = [0.71, 0.75, 0.74, 0.81, 0.80]
bleu = [0.65, 0.69, 0.68, 0.76, 0.74]
f1_score = [0.68, 0.72, 0.71, 0.79, 0.77]
latency = [0.75, 0.80, 1.50, 0.70, 0.65]

# Colors for the bars
colors = ["#FF6B6B", "#4ECDC4", "#45B7D1", "#FF6B6B", "#4ECDC4"]

# Create a 2x2 grid of subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 8))

# Chart 1: ROUGE-L
ax1.bar(methods, rouge_l, color=colors, edgecolor="black")
ax1.set_ylim(0, 1.0)
ax1.set_ylabel("ROUGE-L Score")
ax1.set_title("Average ROUGE-L by Method")
ax1.tick_params(axis='x', rotation=45)

# Chart 2: BLEU
ax2.bar(methods, bleu, color=colors, edgecolor="black")
ax2.set_ylim(0, 1.0)
ax2.set_ylabel("BLEU Score")
ax2.set_title("Average BLEU by Method")
ax2.tick_params(axis='x', rotation=45)

# Chart 3: F1-score
ax3.bar(methods, f1_score, color=colors, edgecolor="black")
ax3.set_ylim(0, 1.0)
ax3.set_ylabel("F1 Score")
ax3.set_title("Average F1-Score by Method")
ax3.tick_params(axis='x', rotation=45)

# Chart 4: Latency
ax4.bar(methods, latency, color=colors, edgecolor="black")
ax4.set_ylim(0, max(latency) + 0.5)
ax4.set_ylabel("Latency (s)")
ax4.set_title("Average Latency by Method")
ax4.tick_params(axis='x', rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the charts
plt.show()
```
